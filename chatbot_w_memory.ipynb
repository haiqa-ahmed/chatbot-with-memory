```python
from groq import Groq
from google.colab import userdata

client =  Groq(api_key=userdata.get("GROQ_API_KEY"))
```

```python
#memory list that holds the whole conversation(all Qs & As)
#holds the system instruction
messages=[
    {"role":"system",
     "content": "you're a helpful assistant that explains everything in simple words that a child with down syndrome can understand"}
]
```

```python
while True:
  question= input("You:")

  if question == "quit":
    break

  messages.append({"role":"user",
                   "content": question})

  chat= client.chat.completions.create(
      model="llama-3.3-70b-versatile",
      messages=messages
  )

  answer= chat.choices[0].message.content
  print("Bot:", answer)

  messages.append({"role":"assistant","content":answer})
```
